```python
import numpy as np

def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def step(x, h, c):

    Wf, Uf, bf = 0.5, 0.1, 0
    Wi, Ui, bi = 0.6, 0.2, 0
    Wc, Uc, bc = 0.7, 0.3, 0
    Wo, Uo, bo = 0.8, 0.4, 0

    f = sigmoid(Wf*x + Uf*h + bf)
    i = sigmoid(Wi*x + Ui*h + bi)

    c_bar = np.tanh(Wc*x + Uc*h + bc)

    c = f*c + i*c_bar

    o = sigmoid(Wo*x + Uo*h + bo)

    h = o * np.tanh(c)

    return h, c, f, i, c_bar, o
```

```python
sequence = [1, 2, 3]

h = 0
c = 0

for t in range(len(sequence)):

    x = sequence[t]

    h, c, f, i, c_bar, o = step(x, h, c)

    print("Step", t+1)
    print("Input:", x)
    print("Forget:", f)
    print("Input gate:", i)
    print("Candidate:", c_bar)
    print("Cell:", c)
    print("Output gate:", o)
    print("Hidden:", h)
    print()
```
